## Question 80
Consider an arbitrary array, write a function that extracts a subpart with a fixed shape and centered on a given element (pad with a fill value when necessary).

```python
def extract_subpart(Z, center, shape, fill=0):
    pad_y = shape[0] // 2
    pad_x = shape[1] // 2
    Zp = np.pad(Z, ((pad_y, pad_y), (pad_x, pad_x)), mode='constant', constant_values=fill)
    y, x = center
    y += pad_y
    x += pad_x
    return Zp[y - pad_y:y + pad_y + 1, x - pad_x:x + pad_x + 1]

Z = np.arange(25).reshape(5, 5)
print(extract_subpart(Z, center=(0, 0), shape=(3, 3), fill=-1))
```

## Question 88
How to implement the Game of Life using numpy arrays?

```python

Z = np.random.randint(0, 2, (5, 5))

# Count live neighbors using wrap-around boundaries
neighbors = (
    np.roll(Z, 1, axis=0) + np.roll(Z, -1, axis=0) +
    np.roll(Z, 1, axis=1) + np.roll(Z, -1, axis=1) +
    np.roll(np.roll(Z, 1, axis=0), 1, axis=1) +
    np.roll(np.roll(Z, 1, axis=0), -1, axis=1) +
    np.roll(np.roll(Z, -1, axis=0), 1, axis=1) +
    np.roll(np.roll(Z, -1, axis=0), -1, axis=1)
)

# Apply Game of Life rules
Z_next = ((neighbors == 3) | ((Z == 1) & (neighbors == 2))).astype(int)

print("Initial Grid:")
print(Z)

print("\nNext Generation:")
print(Z_next)
```

## Question 93

## Question 98
Considering a path described by two vectors (X,Y), how to sample it using equidistant samples?

```python
X = np.array([0, 1, 2, 3, 4])
Y = np.array([0, 1, 1, 2, 2])
dx = np.diff(X)
dy = np.diff(Y)
dist = np.sqrt(dx**2 + dy**2)
t = np.concatenate([[0], np.cumsum(dist)])
t_new = np.linspace(0, t[-1], 10)
X_new = np.interp(t_new, t, X)
Y_new = np.interp(t_new, t, Y)
print(np.column_stack((X_new, Y_new)))
```
