Usefull Function of String

```python
name="divyansh Kumar sinGH"
lenght=len(name)
print("Lenght of the name is:", lenght)
print("Name in uppercase:", name.upper())
print("Name in lowercase:", name.lower())
print("Number of spaces in the name:", name.count(" "))
print("Is all alphabet characters in the name:", name.isalpha())
print(f"Full name without space is {name.replace(' ','')}")
print(f"The first occurance of 'i' is at index {name.find('i')}",f"so place is {float(name.find('i'))+1} ")
```
