# 2. Strings & Conditional Statements

### üéØ Learning Objectives
By the end of this section, you will be able to:
*   Manipulate strings using indexing and slicing.
*   Control program flow with `if`, `elif`, and `else`.

### üë®‚Äçüè´ Tutor's Analogy: "String of Pearls" & "Traffic Cop"
1.  **Strings**: Think of a string as a **Necklace of Pearls**. Each character is a pearl.
    *   **Indexing**: You can point to the 1st pearl (`s[0]`) or the last (`s[-1]`).
    *   **Immutability**: You cannot change a single pearl while it's on the string. You have to restring the whole necklace to make a change.
2.  **Conditionals**: Think of `if/else` as a **Traffic Cop** at a junction.
    *   "IF the light is green, go."
    *   "ELSE IF (elif) the light is yellow, slow down."
    *   "ELSE (otherwise), stop."

### üìù Guided Notes
*   **Indexing:** `s[0]` (first), `s[-1]` (last).
*   **Slicing:** `s[start:stop:step]`.
    *   `s[:3]` -> First 3 chars.
    *   `s[::-1]` -> Reverse the string.
*   **Conditionals Flow:**
    1.  Python checks the `if` condition. If True, executes block and **skips the rest**.
    2.  If False, it checks the first `elif`.
    3.  If none are True, executes `else`.

### üí° Why this matters in real code
Strings are everywhere: **parsing log files**, **validating user passwords**, or **generating reports**. Conditionals are the brain of your program‚Äîthey decide whether to **grant access**, **trigger an alarm**, or **skip a corrupted data row**.

````python
# Example 1: String Methods for Data Cleaning
raw_input = "  user@Example.com  "
clean_email = raw_input.strip().lower() # Remove spaces and convert to lowercase
print(f"Raw: '{raw_input}' -> Clean: '{clean_email}'")

# Example 2: Slicing for Parsing
# Imagine a fixed-format ID: "US-12345-NY"
item_id = "US-12345-NY"
country_code = item_id[:2]      # First 2 chars: "US"
serial_num = item_id[3:8]       # Middle 5 chars: "12345"
print(f"Country: {country_code}, Serial: {serial_num}")

# Example 3: Conditionals in Access Control
user_role = "admin"
is_logged_in = True
has_permission = False

if not is_logged_in:
    print("Please log in.")
elif user_role == "admin":
    print("Welcome, Admin! Full access granted.")
elif has_permission:
    print("Welcome! Standard access granted.")
else:
    print("Access Denied.")

# Example 4: Logical Operators
temperature = 25
is_raining = False

# "Nice weather" if temp is between 20 and 30 AND it's not raining
if 20 <= temperature <= 30 and not is_raining:
    print("It's a nice day for a walk!")
else:
    print("Maybe stay inside.")
````

**Task 3: Password Strength**
*   **Goal:** Return "Weak" (len < 6), "Medium" (6-10), "Strong" (>10).

**Task 4: Extract Domain**
*   **Goal:** Extract text after '@' in an email.
*   **Input:** `"user@example.com"` -> `"example.com"`

**Challenge Task: Login Validator**
*   **Goal:** Return `True` if `username` is "admin" AND `password` is "secret123". Otherwise `False`.

````python
def reverse_str(s):
    # Hint: Use slicing with step -1
    return s[::-1]

def count_vowels(s):
    # Hint: Iterate chars or use .count() for each vowel
    s = s.lower()
    return sum(s.count(v) for v in 'aeiou')

def password_strength(pw):
    # Hint: Use len(pw) and if/elif/else
    l = len(pw)
    if l < 6:
        return "Weak"
    elif l <= 10:
        return "Medium"
    else:
        return "Strong"

def extract_domain(email):
    # Hint: Use .split('@') and take the second part
    return email.split('@')[-1] if '@' in email else ''

def login_check(username, password):
    # Hint: Return True only when username=="admin" and password=="secret123"
    return username == 'admin' and password == 'secret123'
````

*   **Dynamic Arrays**: Python lists are implemented as dynamic arrays.
*   **Time Complexity**:
    *   Access (`arr[i]`): **O(1)**
    *   Append (`arr.append(x)`): **O(1)** (amortized)

### üß† Guided Exercises (Lists & Tuples)

**Task 1: Sum List**

**Task 2: Remove Duplicates**
*   **Goal:** Return a new list with unique elements, preserving order.
*   **Input:** `[1, 2, 2, 3, 1]` -> `[1, 2, 3]`

**Task 3: Merge Sorted Lists**
*   **Goal:** Combine two lists and return a sorted result.
*   **Input:** `[1, 3]`, `[2, 4]` -> `[1, 2, 3, 4]`

**Challenge Task: Second Largest**
*   **Goal:** Find the second largest number in a list *without* using `sort()`.
*   **Input:** `[10, 20, 4, 45, 99]` -> `45`

````python
def word_freq(text):
    # Hint: Split text and loop, using d[word] = d.get(word, 0) + 1
    d = {}
    for word in text.split():
        d[word] = d.get(word, 0) + 1
    return d

def common_elements(l1, l2):
    # Hint: Convert both to sets and use & operator or .intersection()
    return set(l1) & set(l2)

def invert_dict(d):
    # Hint: Loop d.items() and swap
    return {v: k for k, v in d.items()}

def high_scorers(scores):
    # Hint: Loop d.items() and check if value > 80
    return [name for name, score in scores.items() if score > 80]
````

````python
def sum_list(lst):
    """Return sum of all numbers in a list."""
    return sum(lst)

def remove_duplicates(lst):
    """Return new list with unique elements preserving order."""
    seen = set()
    res = []
    for x in lst:
        if x not in seen:
            seen.add(x)
            res.append(x)
    return res

def merge_sorted(l1, l2):
    """Merge two sorted lists into a sorted list."""
    i = j = 0
    out = []
    while i < len(l1) and j < len(l2):
        if l1[i] <= l2[j]:
            out.append(l1[i]); i += 1
        else:
            out.append(l2[j]); j += 1
    out.extend(l1[i:])
    out.extend(l2[j:])
    return out

def second_largest(lst):
    """Find second largest without sorting."""
    if len(lst) < 2:
        return None
    first = second = float('-inf')
    for x in lst:
        if x > first:
            second = first
            first = x
        elif x > second and x != first:
            second = x
    return second if second != float('-inf') else None
````

````python
# üß™ Test Your Code
try:
    assert sum_list([1, 2, 3]) == 6
    assert remove_duplicates([1, 2, 2, 3, 1]) == [1, 2, 3]
    assert merge_sorted([1, 3], [2, 4]) == [1, 2, 3, 4]
    
    # Challenge Test
    assert second_largest([10, 20, 4, 45, 99]) == 45
    
    print("‚úÖ All Lists & Tuples tests passed!")
except AssertionError as e:
    print(f"‚ùå Test Failed: {e}")
````

# 4. Dictionary & Set

### üéØ Learning Objectives
By the end of this section, you will be able to:
*   Store key-value pairs in Dictionaries.
*   Store unique elements in Sets.
*   Perform set operations (union, intersection).
*   Choose the right data structure for the job.

### üì∫ Lesson Anchor
*   **Video:** Shradha Khapra ‚Äî Dictionary & Set
*   **Summary:** Covers `dict` key‚Üívalue mappings and `set` for unique elements. Explains creating, accessing, updating, deleting keys, `.get()` with default, `.setdefault()`, iteration over keys/values/items.

### üìù Guided Notes
*   **Dictionary:** `{"key": "value"}`. Fast lookups O(1). Keys must be immutable.
*   **Set:** `{1, 2, 3}`. Unique elements only. No duplicates. Unordered.
*   **Methods:** `d.get(key, default)`, `d.keys()`, `d.values()`, `d.items()`. `s.add(x)`, `s.remove(x)`, `s.union(s2)`, `s.intersection(s2)`.
*   **When to use what?**
    *   **List:** When order matters and duplicates are allowed. (e.g., a playlist of songs).
    *   **Dictionary:** When you need to look up values by a unique key. (e.g., phonebook: name -> number).
    *   **Set:** When you need uniqueness and fast membership testing (checking `if x in s`). (e.g., list of active user IDs).

### üí° Why this matters in real code
**Dictionaries** are the database of your code. You'll use them to represent **user profiles**, **configuration settings**, or **JSON responses** from an API. **Sets** are essential for **deduplication** (e.g., "how many *unique* visitors did we have?") and fast lookups.

### üíª Code Examples

````python
# Example 1: User Profile (Dictionary)
user = {
    "id": 101,
    "username": "coder_123",
    "is_admin": False
}
print(f"User: {user['username']}")
print(f"Role: {user.get('role', 'Guest')}") # Safe access with default

# Example 2: Counting Frequency (Dict Pattern)
text = "apple banana apple orange banana apple"
counts = {}
for word in text.split():
    # Get current count (0 if new) and add 1
    counts[word] = counts.get(word, 0) + 1
print(f"Word Counts: {counts}")

# Example 3: Unique Visitors (Set)
visitors_monday = {"alice", "bob", "charlie"}
visitors_tuesday = {"bob", "david", "alice"}

# Union: All unique visitors
all_visitors = visitors_monday.union(visitors_tuesday)
print(f"All Visitors: {all_visitors}")

# Intersection: Who came both days?
loyal_visitors = visitors_monday.intersection(visitors_tuesday)
print(f"Loyal Visitors: {loyal_visitors}")
````

### üß† Guided Exercises (Dictionary & Set)

**Task 1: Word Frequency**
*   **Goal:** Return a dictionary where keys are words and values are counts.
*   **Input:** `"a b a"` -> `{"a": 2, "b": 1}`

**Task 2: Common Elements**
*   **Goal:** Return a set of elements present in both lists.
*   **Input:** `[1, 2, 3]`, `[2, 3, 4]` -> `{2, 3}`

**Task 3: Invert Dictionary**
*   **Goal:** Swap keys and values. Assume values are unique.
*   **Input:** `{"a": 1, "b": 2}` -> `{1: "a", 2: "b"}`

**Challenge Task: Student Grades**
*   **Goal:** Given `{"Alice": 85, "Bob": 92, "Charlie": 78}`, return a list of names who scored > 80.
*   **Output:** `["Alice", "Bob"]` (Order doesn't matter)

````python
def word_freq(text):
    # Hint: Split text and loop, using d[word] = d.get(word, 0) + 1
    d = {}
    for word in text.split():
        d[word] = d.get(word, 0) + 1
    return d

def common_elements(l1, l2):
    # Hint: Convert both to sets and use & operator or .intersection()
    return set(l1) & set(l2)

def invert_dict(d):
    # Hint: Loop d.items() and swap
    return {v: k for k, v in d.items()}

def high_scorers(scores):
    # Hint: Loop d.items() and check if value > 80
    return [name for name, score in scores.items() if score > 80]
````

````python
# üöÄ DSA Pattern: Prefix Sum (using Loops)
# Problem: Calculate sum of elements between indices i and j efficiently.
nums = [1, 2, 3, 4, 5]
prefix_sum = [0] * (len(nums) + 1)

for i in range(len(nums)):
    prefix_sum[i+1] = prefix_sum[i] + nums[i]

print(f"Original: {nums}")
print(f"Prefix Sum: {prefix_sum}")
# Sum from index 1 to 3 (2+3+4 = 9) -> prefix_sum[4] - prefix_sum[1]
print(f"Sum(1,3): {prefix_sum[4] - prefix_sum[1]}")

# üöÄ DSA Pattern: Sliding Window (Fixed Size)
# Problem: Find max sum of subarray of size k
k = 3
current_sum = sum(nums[:k])
max_sum = current_sum

for i in range(len(nums) - k):
    current_sum = current_sum - nums[i] + nums[i+k]
    max_sum = max(max_sum, current_sum)

print(f"Max sum of size {k}: {max_sum}")
````

````python
def sum_n(n):
    # Hint: Use range(1, n+1)
    return sum(range(1, n + 1))

def factorial(n):
    # Hint: Initialize result=1 and multiply in loop
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def count_digits(n):
    # Hint: Convert to string OR use while n > 0: n //= 10
    return len(str(abs(n)))

def is_prime(n):
    # Hint: Handle n < 2. Loop range(2, int(n**0.5) + 1)
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def star_pattern(n):
    # Hint: Loop i from 1 to n. Append "*" * i to list.
    return ["*" * i for i in range(1, n + 1)]
````

````python
# üöÄ DSA Pattern: Simple DFS on a Grid (Recursion)
def dfs(grid, r, c, visited):
    rows, cols = len(grid), len(grid[0])
    
    # Base cases: out of bounds, visited, or blocked (0)
    if (r < 0 or r >= rows or c < 0 or c >= cols or 
        (r, c) in visited or grid[r][c] == 0):
        return
    
    visited.add((r, c))
    print(f"Visiting ({r}, {c})")
    
    # Explore neighbors (Up, Down, Left, Right)
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    for dr, dc in directions:
        dfs(grid, r + dr, c + dc, visited)

# 1 = Land, 0 = Water
grid_map = [
    [1, 1, 0],
    [0, 1, 0],
    [0, 1, 1]
]
visited_set = set()
print("Starting DFS from (0,0):")
dfs(grid_map, 0, 0, visited_set)
````

### üìù Guided Notes
*   **Definition:** `def func_name(param):`.
*   **Return:** `return value` sends result back.
*   **Recursion:** Base case (stop condition) + Recursive step.
    *   **Dry Run `fact(3)`:**
        1.  `fact(3)` calls `3 * fact(2)`
        2.  `fact(2)` calls `2 * fact(1)`
        3.  `fact(1)` returns `1` (Base Case)
        4.  `fact(2)` returns `2 * 1 = 2`
        5.  `fact(3)` returns `3 * 2 = 6`
*   **Time Complexity Note:** Naive recursive Fibonacci is very slow (Exponential Time $O(2^n)$) because it re-calculates the same values many times. Iterative or Memoized versions are much faster (Linear Time $O(n)$).
*   **Scope:** Variables inside function are local.

### üí° Why this matters in real code
Functions allow you to **write once, use everywhere**. Instead of copying the same 10 lines of code to calculate tax in 5 different places, you write one `calculate_tax()` function. This makes your code **modular**, **testable**, and **easier to fix**.

### üíª Code Examples

````python
# Example 1: Reusable Logic
def calculate_total(price, tax_rate=0.05):
    """Calculates total price including tax."""
    tax = price * tax_rate
    return price + tax

print(f"Total: {calculate_total(100)}")      # Uses default tax 0.05
print(f"Total: {calculate_total(100, 0.1)}") # Uses custom tax 0.1

# Example 2: Returning Multiple Values
def get_stats(numbers):
    return min(numbers), max(numbers), sum(numbers)

low, high, total = get_stats([10, 20, 30])
print(f"Min: {low}, Max: {high}, Sum: {total}")

# Example 3: Recursion (Factorial)
def fact(n):
    # 1. Base Case: Stop when n is 0 or 1
    if n <= 1:
        return 1
    # 2. Recursive Step: n * fact(n-1)
    return n * fact(n - 1)

print(f"5! = {fact(5)}")
````

````python
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []
    
    def add_grade(self, grade):
        self.grades.append(grade)
    
    def average(self):
        return sum(self.grades) / len(self.grades) if self.grades else 0

class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    
    def area(self):
        return self.w * self.h

class Cart:
    def __init__(self):
        self.items = []
        
    def add_item(self, name, price):
        self.items.append({'name': name, 'price': price})
        
    def total_cost(self):
        return sum(item['price'] for item in self.items)
````

````python
import os
import math
from datetime import datetime

def write_log(message):
    # Hint: with open('log.txt', 'a') as f: ...
    with open('log.txt', 'a') as f:
        f.write(f"{datetime.now()}: {message}\n")

def circle_stats(radius):
    # Hint: area = pi * r^2, circ = 2 * pi * r
    area = math.pi * (radius ** 2)
    circ = 2 * math.pi * radius
    return area, circ

def file_stats(filename):
    # Hint: f.readlines() gives lines. split() gives words.
    try:
        with open(filename, 'r') as f:
            lines = f.readlines()
            num_lines = len(lines)
            num_words = sum(len(line.split()) for line in lines)
            return num_lines, num_words
    except FileNotFoundError:
        return 0, 0
````

````python
import os
import csv

# --- Your Code Here ---

class Transaction:
    def __init__(self, date, category, amount, description):
        self.date = date
        self.category = category
        self.amount = float(amount)
        self.description = description
        
    def to_list(self):
        return [self.date, self.category, self.amount, self.description]

class BudgetManager:
    def __init__(self, filename='budget.csv'):
        self.filename = filename
        self.transactions = []
        self.load_data()
        
    def load_data(self):
        self.transactions = []
        if not os.path.exists(self.filename):
            return
            
        try:
            with open(self.filename, 'r', newline='') as f:
                reader = csv.reader(f)
                header = next(reader, None) # Skip header
                for row in reader:
                    if row:
                        self.transactions.append(Transaction(row[0], row[1], row[2], row[3]))
        except (FileNotFoundError, ValueError):
            pass # Handle empty files or bad data gracefully
        
    def save_data(self):
        with open(self.filename, 'w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(['Date', 'Category', 'Amount', 'Description'])
            for t in self.transactions:
                writer.writerow(t.to_list())
        
    def add_transaction(self, date, category, amount, desc):
        t = Transaction(date, category, amount, desc)
        self.transactions.append(t)
        self.save_data()
        
    def get_summary(self):
        summary = {}
        for t in self.transactions:
            summary[t.category] = summary.get(t.category, 0) + t.amount
        return summary

# --- End Your Code ---

# üß™ Test Logic
try:
    test_file = 'test_budget.csv'
    if os.path.exists(test_file): os.remove(test_file)
    
    manager = BudgetManager(test_file)
    manager.add_transaction("2023-01-01", "Food", 50, "Groceries")
    manager.add_transaction("2023-01-02", "Food", 20, "Snacks")
    manager.add_transaction("2023-01-03", "Transport", 30, "Bus")
    
    # Check Persistence
    manager2 = BudgetManager(test_file)
    assert len(manager2.transactions) == 3
    
    # Check Summary
    summary = manager.get_summary()
    assert summary["Food"] == 70
    assert summary["Transport"] == 30
    
    print("‚úÖ Capstone Logic Verified!")
except Exception as e:
    print(f"‚ùå Capstone Test Failed: {e}")
````

````python
from collections import deque, defaultdict, Counter

# 1. Deque (Double-ended Queue) - O(1) append/pop from both ends
queue = deque([1, 2, 3])
queue.appendleft(0)
queue.append(4)
print(f"Deque: {queue}")
queue.popleft() # Removes 0

# 2. Defaultdict - Handles missing keys automatically
graph = defaultdict(list)
graph['A'].append('B') # No KeyError!
print(f"Graph: {dict(graph)}")

# 3. Counter - Frequency map
nums = [1, 1, 2, 3, 3, 3, 4]
counts = Counter(nums)
print(f"Counts: {counts}")
print(f"Most common: {counts.most_common(1)}")
````

````python
import heapq
import bisect

# 1. Heapq (Min-Heap) - Priority Queue
min_heap = []
heapq.heappush(min_heap, 5)
heapq.heappush(min_heap, 1)
heapq.heappush(min_heap, 3)
print(f"Heap: {min_heap}") # Smallest is always at index 0
print(f"Pop smallest: {heapq.heappop(min_heap)}")

# 2. Bisect (Binary Search)
sorted_arr = [1, 3, 5, 7, 9]
# Find insertion point to maintain order
idx = bisect.bisect_left(sorted_arr, 6)
print(f"Insert 6 at index: {idx}")
bisect.insort(sorted_arr, 6)
print(f"Sorted: {sorted_arr}")
````

````python
# üöÄ Algorithmic Pattern: Binary Search (Iterative)
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
            
    return -1

arr = [10, 20, 30, 40, 50]
print(f"Found 30 at index: {binary_search(arr, 30)}")
print(f"Found 25 at index: {binary_search(arr, 25)}")
````

````python
# üöÄ Algorithmic Pattern: Backtracking Template (Subsets)
def generate_subsets(nums):
    result = []
    
    def backtrack(start, current_subset):
        # Add copy of current subset to results
        result.append(current_subset[:])
        
        for i in range(start, len(nums)):
            # Include nums[i]
            current_subset.append(nums[i])
            # Move to next element
            backtrack(i + 1, current_subset)
            # Backtrack: remove nums[i]
            current_subset.pop()
            
    backtrack(0, [])
    return result

print(f"Subsets of [1, 2]: {generate_subsets([1, 2])}")
````

## Phase 5: Time Series Forecasting
**Goal**: Predict trends over time (Stock prices, Traffic flow).

### üë®‚Äçüè´ Tutor's Analogy: "Weather Forecasting"
*   **Trend**: Is the temperature generally getting hotter (Summer is coming)?
*   **Seasonality**: Does it get cold *every* night and hot *every* day? (Repeating patterns).
*   **Noise**: Did it rain randomly today? (Random fluctuations).
*   **Forecasting**: Using the Trend + Seasonality to guess tomorrow's temperature, ignoring the random noise.

### üîë Key Concepts
*   **Time Index**: Data must be indexed by date (not 0, 1, 2...).
*   **Seasonality**: Repeating patterns (e.g., ice cream sales peak in summer).
*   **Rolling Average**: Smoothing out "noise" to see the long-term trend.

*(Note: For advanced forecasting, libraries like `Prophet` or `Statsmodels` are used. Below is a fundamental Pandas approach).*

````python
# Generating a Time Series
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100).cumsum() # Random walk

ts_df = pd.DataFrame({'Date': dates, 'Value': values})
ts_df.set_index('Date', inplace=True)

# Rolling Average (Smoothing)
# This calculates the average of the last 7 days to show the trend
ts_df['7-Day Moving Avg'] = ts_df['Value'].rolling(window=7).mean()

# Plotting with Pandas (uses Matplotlib backend)
ts_df.plot(title="Time Series with Moving Average", figsize=(10, 5))
plt.show()

# üß† Study Question: What happens if you increase the window to 30?
# Answer: The line becomes smoother but reacts slower to changes.
````
