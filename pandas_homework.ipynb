```python
# Question 25: Create a Seniority column based on Experience
# >= 10 years = Senior, 5-9 years = Mid, < 5 years = Junior
print("Question 25: Add Seniority column based on Experience")

# Initialize Seniority column
df['Seniority'] = ''

# Add values based on Experience
for index in range(len(df)):
    if df.loc[index, 'Experience'] >= 10:
        df.loc[index, 'Seniority'] = 'Senior'
    elif df.loc[index, 'Experience'] >= 5:
        df.loc[index, 'Seniority'] = 'Mid'
    else:
        df.loc[index, 'Seniority'] = 'Junior'

print(df[['Name', 'Experience', 'Seniority']].head(10))
```

```python
# Question 35: Change bar color to green and add Salary values on top of bars
print("Question 35: Bar chart with green color and salary values on top")
plt.figure(figsize=(14, 6))
bars = plt.bar(df['Name'], df['Salary'], color='green', alpha=0.7)

# Add salary values on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom', fontsize=8)

plt.xlabel('Name', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.title('Salary by Employee', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Question 38: Scatter plot of Age vs Salary
print("Question 38: Scatter plot of Age vs Salary")

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Salary'], alpha=0.6, s=100)
plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.title('Age vs Salary', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Question 39: Color scatter points by Department
print("Question 39: Scatter plot of Age vs Salary (colored by Department)")

# Define colors for each department
dept_colors = {'IT': 'red', 'HR': 'blue', 'Finance': 'green'}
colors = df['Department'].map(dept_colors)

plt.figure(figsize=(10, 6))
for dept in df['Department'].unique():
    dept_data = df[df['Department'] == dept]
    plt.scatter(dept_data['Age'], dept_data['Salary'], 
                label=dept, alpha=0.6, s=100)

plt.xlabel('Age (years)', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.title('Age vs Salary (colored by Department)', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Question 40: Scatter plot of Experience vs Salary with regression line
print("Question 40: Scatter plot of Experience vs Salary with regression line")

# Calculate the regression line
# Simple linear regression: y = mx + b
x = df['Experience'].values
y = df['Salary'].values

# Calculate slope (m) and intercept (b)
m = np.polyfit(x, y, 1)[0]
b = np.polyfit(x, y, 1)[1]

# Create regression line points
x_line = np.array([x.min(), x.max()])
y_line = m * x_line + b

plt.figure(figsize=(10, 6))
plt.scatter(df['Experience'], df['Salary'], alpha=0.6, s=100, label='Employees')
plt.plot(x_line, y_line, color='red', linewidth=2, label='Trend Line')
plt.xlabel('Experience (years)', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.title('Experience vs Salary (with Trend Line)', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Regression equation: Salary = {m:.2f} * Experience + {b:.2f}")
```
